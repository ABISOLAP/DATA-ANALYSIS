```python
pre=df.groupby('Region')['Total Profit'].sum().sort_values(ascending=False)# Total Profit per Region from highest to lowest
pre
```

```python
pre.plot(kind='barh')
plt.xlabel("Total Profit")
plt.ylabel("Region")
plt.title("Total Profit per region")
```
